```
# to check given number is even or odd
n= int(input("enter the number"))
if n%2 == 0:
 print(n,"is a even number")
else:
  print(n,"is odd number")

```

```
# to check given number is perfectly multiple of 3 and 5
n=int(input("enter the number"))

if n%3 == 0 and n%5 == 0:
   print(n,"the given number is  perfect multiple of 3 and 5")              
else:
   print(n,"not a multiple of 3 and 5")  
```

```
 # to check given number is positive,negative or zero
 n=int(input("enter the number"))
if n>0:
   print(n,"it is a positive number")
elif n<0:
   print(n,"it is a negative number")
else:
   print(n,"it is a zero")
```

```
# to check the greatest of three numbers
n1=int(input("enter the first number"))
n2=int(input("enter the second number"))
n3=int(input("enter the third number"))
if n1>n2 and n1>n3:
   print(n1,"is the greatest number")
elif n2>n1 and n2>n3:
   print(n2,"is the greatest number")
else:
   print(n3,"is the greatest number")   
```

```
# given year is leap year or not
n=int(input("enter the year "))
if n%4== 0 and n%100==0 and n%400==0:
  print(n,"is a leap year")
else:
  print(n,"is not a leap year")
```
